```python
import numpy as np

import dlib

import matplotlib.pyplot as plt
# 魔法指令，快速显示图片
%matplotlib inline
```

```python
# 图片通道：红绿蓝
jin = plt.imread('./jin.jpg')

plt.imshow(jin)
```

```python
# cv2可以对图片进行人脸识别
# cv2识别人脸时，需要数据，haar类型数据，人脸的特征
import cv2
```

```python
# cv2 ------>图片通道：蓝绿红
cv2.imshow('jin',jin[:,:,::-1])

# 等待键盘的输入
cv2.waitKey(0)

# 释放资源
cv2.destroyAllWindows()
```

```python
# dlib人脸识别

face_detector = dlib.get_frontal_face_detector()
```

```python
# 识别人脸时，无论是彩色还是黑白，都可以识别
# 黑白，内存小，速度快
# 彩色变黑白，尺寸完全一样
jin = cv2.imread('./jin.jpg')

# 蓝绿红
# jin = jin[:,:,::-1]

gray = cv2.cvtColor(jin,code = cv2.COLOR_BGR2GRAY)

faces = face_detector(gray,1)

for face in faces:
    top = face.top()
    left = face.left()
    bottom = face.bottom()
    right = face.right()
    print(top,left,bottom,right)
#     cv2将识别出来的人脸，进行绘制
    cv2.rectangle(jin,pt1 = (left,top),pt2 = (right,bottom),color = [0,0,255],thickness = 2)
    
cv2.imshow('face',jin)

cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
# 识别人脸时，无论是彩色还是黑白，都可以识别
# 黑白，内存小，速度快
# 彩色变黑白，尺寸完全一样
image = cv2.imread('./tuan.jpg')

gray = cv2.cvtColor(image,code = cv2.COLOR_BGR2GRAY)

faces = face_detector(gray,1)

for face in faces:
    top = face.top()
    left = face.left()
    bottom = face.bottom()
    right = face.right()
    print(top,left,bottom,right)
#     cv2将识别出来的人脸，进行绘制
    cv2.rectangle(image,pt1 = (left,top),pt2 = (right,bottom),color = [0,0,255],thickness = 2)
    
cv2.imshow('face',image)

cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
# 识别人脸时，无论是彩色还是黑白，都可以识别
# 黑白，内存小，速度快
# 彩色变黑白，尺寸完全一样
image = cv2.imread('./nba3.jpg')

gray = cv2.cvtColor(image,code = cv2.COLOR_BGR2GRAY)

faces = face_detector(gray,3)

for face in faces:
    top = face.top()
    left = face.left()
    bottom = face.bottom()
    right = face.right()
    print(top,left,bottom,right)
#     cv2将识别出来的人脸，进行绘制
    cv2.rectangle(image,pt1 = (left,top),pt2 = (right,bottom),color = [0,255,0],thickness = 2)
    
cv2.imshow('face',image)

cv2.waitKey(0)

cv2.destroyAllWindows()
```
